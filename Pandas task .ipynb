```python
import pandas as pd
import numpy as np

file_path = 'C:/Users/Maks/Desktop/Energy Indicators.xls'

energy = pd.read_excel(file_path, skiprows=17, header=None, skipfooter=38)
energy = energy.drop([0, 1], axis=1)
energy.columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
energy['Energy Supply'] = energy['Energy Supply'] * 1000000
energy = energy.replace('...', np.NaN)
energy['Country'] = energy['Country'].replace({'Republic of Korea': 'South Korea',
'United States of America': 'United States',
'United Kingdom of Great Britain and Northern Ireland': 'United Kingdom',
'China': 'Hong Kong', 'Hong Kong Special Administrative Region': 'Hong Kong'})
energy['Country'] = energy['Country'].str.replace(r'\(.*\)', '')  # Видалення дужок
energy['Country'] = energy['Country'].str.replace(r'\d+', '')  # Видалення чисел

energy


```

```python
file_path = 'C:/Users/Maks/Desktop/API_NY.GDP.MKTP.CD_DS2_en_csv_v2_5871885.csv'

GDP = pd.read_csv(file_path, header=2)
GDP = GDP.rename(columns={'Country Name': 'Country'})
GDP['Country'] = GDP['Country'].replace({
    'Korea, Rep.': 'South Korea',
    'Iran, Islamic Rep.': 'Iran',
    'Hong Kong SAR, China': 'Hong Kong'
})
GDP
```

```python
file_path = 'C:/Users/Maks/Desktop/scimagojr.xlsx'

ScimEn = pd.read_excel(file_path)
ScimEn = ScimEn.reset_index(drop=True)

ScimEn
```

```python
merged = pd.merge(ScimEn, energy, on='Country', how='outer')
merged = pd.merge(merged, GDP, on='Country', how='outer')
GDP = GDP[['Country', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']]
answer_one = pd.merge(ScimEn, energy, how='outer', left_on='Country', right_on='Country')
answer_one = pd.merge(answer_one, GDP, how='outer', left_on='Country', right_on='Country')
answer_one.set_index('Country', inplace=True)
answer_one = answer_one[answer_one['Rank'] <= 15]
answer_one['Rank'] = answer_one['Rank'].astype(int)
answer_one['Documents'] = answer_one['Documents'].astype(int)
answer_one['Citable documents'] = answer_one['Citable documents'].astype(int)
answer_one['Citations'] = answer_one['Citations'].astype(int)
answer_one['Self-citations'] = answer_one['Self-citations'].astype(int)
answer_one['H index'] = answer_one['H index'].astype(int)

columns_to_keep = ['Rank', 'Documents', 'Citable documents', 
                   'Citations', 'Self-citations',
                   'Citations per document', 
                   'H index', 
                   'Energy Supply',
                   'Energy Supply per Capita',
                   '% Renewable',
                   '2006', 
                   '2007',
                   '2008',
                   '2009',
                   '2010',
                   '2011',
                   '2012',
                   '2013',
                   '2014',
                   '2015']
answer_one = answer_one.loc[:, columns_to_keep]
answer_one

```

```python
def answer_two():
    Top15 = answer_one
    years = ['2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']
    Top15['avgGDP'] = Top15[years].mean(axis=1)
    avgGDP = Top15['avgGDP'].sort_values(ascending=False)
    return avgGDP

print(answer_two())
```

```python
def answer_three():
    Top15 = answer_one
    sixth_largest = Top15.iloc[5]
    gdp_difference = sixth_largest['2015'] - sixth_largest['2006']
    return gdp_difference
print(answer_three())
```

```python
def answer_four():
    Top15 = answer_one
    Top15['Citation_ratio'] = Top15['Self-citations'] / Top15['Citations']
    max_ratio_country = Top15['Citation_ratio'].idxmax()
    max_ratio = Top15['Citation_ratio'].max()
    return (max_ratio_country, max_ratio)
print(answer_four())
```

```python
def answer_five():
    Top15 = answer_one
    Top15['Population'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    Top15.sort_values(by='Population', ascending=False, inplace=True)
    third_most_populous_country = Top15.iloc[2].name
    return third_most_populous_country
print(answer_five())
```

```python
def answer_six():
    Top15 = answer_one
    Top15['Population'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    Top15['Citable docs per Capita'] = Top15['Citable documents'] / Top15['Population']
    df_new = Top15[['Citable docs per Capita', 'Energy Supply per Capita']]
    correlation = df_new.corr(method='pearson')
    return correlation.loc['Citable docs per Capita', 'Energy Supply per Capita']

print(answer_six())

```

```python
def answer_seven():
    Top15 = answer_one
    Top15['Population Estimate'] = (Top15['Energy Supply'] / Top15['Energy Supply per Capita']).astype(float)
    Top15['Continent'] = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}
    
    result = Top15.groupby('Continent')['Population Estimate'].agg(['size', 'sum', 'mean', 'std'])
    
    return result
print(answer_seven())
```
